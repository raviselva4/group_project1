```python
# create df for unique parties 
democrat_2008 = election_data_2008.loc[election_data_2008["party"] == 'democrat']
democrat_2008
```

```python
republican_2008 = election_data_2008.loc[election_data_2008["party"] == 'republican']
republican_2008
```

```python
# collecting required columns...
republican_2008 = republican_2008[['total_votes', 'candidate_votes', 'Location']]
republican_2008.columns = ['total_votes', '2008_republican_votes', 'Location']
republican_2008.head()
democrat_2008 = democrat_2008[['total_votes', 'candidate_votes', 'Location']]
democrat_2008.columns = ['total_votes', '2008_democrat_votes', 'Location']
democrat_2008.head()
```

```python
# Finding # of rows in both df's to make sure sum match with parent df
print(len(republican_2008['Location'].unique()), len(democrat_2008['Location'].unique()))
```

```python
# Finding duplicate rows by Location
dup_df = republican_2008[republican_2008['Location'].duplicated()]
dup_df
```

```python
# pulling duplicate rows for Location
t = republican_2008.loc[(republican_2008['Location'] == 'Virginia, Bedford') | (republican_2008['Location'] == 'Virginia, Fairfax') | (republican_2008['Location'] == 'Virginia, Franklin') | (republican_2008['Location'] == 'Virginia, Richmond') | (republican_2008['Location'] == 'Virginia, Roanoke')]
t1 = t.sort_values(['Location'])
t1
```

```python
# Data cleaning - deleting rows that are not supposed to be counties.. (did a manual check)
ind = republican_2008.loc[((republican_2008['Location'] == 'Virginia, Bedford') & (republican_2008['total_votes'] == 2734) ) | ((republican_2008['Location'] == 'Virginia, Fairfax') & (republican_2008['total_votes'] == 11398) )  | ((republican_2008['Location'] == 'Virginia, Franklin') & (republican_2008['total_votes'] == 4427) ) | ((republican_2008['Location'] == 'Virginia, Richmond') & (republican_2008['total_votes'] == 93085) ) | ((republican_2008['Location'] == 'Virginia, Roanoke') & (republican_2008['total_votes'] == 40772) )].index
ind      # make sure before deletion...
republican_2008.drop(ind, inplace=True)
ind = democrat_2008.loc[((democrat_2008['Location'] == 'Virginia, Bedford') & (democrat_2008['total_votes'] == 2734) ) | ((democrat_2008['Location'] == 'Virginia, Fairfax') & (democrat_2008['total_votes'] == 11398) )  | ((democrat_2008['Location'] == 'Virginia, Franklin') & (democrat_2008['total_votes'] == 4427) ) | ((democrat_2008['Location'] == 'Virginia, Richmond') & (democrat_2008['total_votes'] == 93085) ) | ((democrat_2008['Location'] == 'Virginia, Roanoke') & (democrat_2008['total_votes'] == 40772) )].index
democrat_2008.drop(ind, inplace=True)
```

```python
election_parties_2008 = pd.merge(republican_2008,democrat_2008, on='Location')
election_parties_2008 = election_parties_2008[['Location','total_votes_x','2008_republican_votes','2008_democrat_votes']]
election_parties_2008.columns = ['Location','total_votes','2008_republican_votes','2008_democrat_votes']
election_parties_2008
```

```python
# identifiy mismatch records ....and review before deletion
pop1 = election_parties_2008.merge(temp_df, on='Location', how='left', indicator=True)
pop1
```
